## Run SDP with uv (outside Databricks)

Run Spark Declarative Pipelines on command line, outside Databricks, using [uvx](https://docs.astral.sh/uv/guides/tools/).

<br>

```sh
uvx --with "pyspark[pipelines]==4.1.0" spark-pipelines --help
```

I learnt it while reviewing [Installation requirements for using the create-dagster CLI](https://docs.dagster.io/getting-started/installation#installation-requirements-for-using-the-create-dagster-cli).
