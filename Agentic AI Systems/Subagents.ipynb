# Subagents


1. Agents can delegate tasks to independent subagents that work in the foreground or background.
1. Agents are independent workers to handle subtasks.
1. A subagent shares tools and codebase context with the parent, but operates in its own conversation chain and does not inherit the parent's conversation history.
1. Useful for tasks that benefit from focused, independent work.
    1. Eploring a codebase
    1. Running tests
    1. Implementing a feature in parallel.
1. Agents can be requested to use subagents explicitly (e.g. "research how auth works in a subagent"), or the agent may decide to delegate autonomously when it determines a task would benefit from independent work.
1. Subagents can both improve overall coding performance and reduce cost.

## Subagent profiles

[Execution modes](https://docs.devin.ai/cli/subagents#how-subagents-work):
1. Agents can spawn subagents in one of the available **subagent profiles** and chooses whether the subagent should run in the **foreground** or **background**.
1. In **Foreground** mode, subagents run inline in the parent agent's session.
    1. **Foreground subagents**
    1. The parent agent pauses and waits for the subagent to finish before continuing.
    1. You can approve or deny tool calls as they come up.
    1. Behave like the main agent — you are prompted to approve or deny tool calls as usual.
1. In **Background** mode, subagents run in parallel while the parent agent continues working.
    1. **Background subagents**
    1. The parent is automatically notified when the subagent completes.
    1. Inherit any tool permissions you have already granted during the current session.
    1. Unapproved tools are automatically denied. Any tool that has not been pre-approved is automatically denied.
    1. Background subagents cannot prompt you for new permissions.
    1. If a background subagent fails because a required tool was denied, you can resume it in the foreground to approve the necessary permissions.
1. When a subagent finishes, the parent agent reads the result and summarizes the key findings and actions for you.

[Subagent Profiles](https://docs.devin.ai/cli/subagents#subagent-profiles):
1. Subagents run with a specific profile that determines its capabilities.
1. The agent automatically chooses the appropriate profile based on the task.


## Subagent Cost

[Subagent Cost](https://docs.devin.ai/cli/subagents#subagent-cost):
1. Subagents run as their own agent sessions, each with its own context window and inference calls, so they consume cost independently of the parent.
1. Tasks that spawn multiple subagents (or nest them) consume more credits.
1. Use subagents deliberately when the parallelism or focused context is worth the additional spend.
