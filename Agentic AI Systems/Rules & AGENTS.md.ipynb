# Rules & AGENTS.md


## Rules

1. **Rules** are persistent instructions
1. Provide always-on instructions and context that guide the agent in every session
1. Injected into the agent's context at the start of every session for consistent behavior
1. Use cases: coding standards, architectural guidelines, preferred libraries, testing conventions, and project-specific constraints
1. Rules and AGENTS should be kept as small as possible.
1. Recommended pattern: Use a rule to reference skills that a model should use in particular scenarios.
1. [Add rules with an `AGENTS.md` file in your project root](https://docs.devin.ai/cli/extensibility/rules#agents-md)


### 🚨 Use Skills instead whenever possible

It is highly recommended to use skills instead whenever possible for improved coding ability and lower cost.

Skills are only injected into the context when relevant.

## Global Rules

[Global Rules](https://docs.devin.ai/cli/extensibility/rules#global-rules):

1. Loaded at the start of every session, regardless of which project you're working in.
1. Used for personal preferences that apply everywhere.
1. Work alongside project rules — both are loaded and active at the same time.
    1. `~/.claude/CLAUDE.md` for [Claude Code](https://claude.com/product/claude-code)
    1. `~/.config/devin/AGENTS.md` for [Devin CLI](https://devin.ai/cli)

## AGENTS.md for Always-On Rules

`AGENTS.md` is the recommended approach for project rules.

Easy to read, version-controlled, and works across multiple AI tools.

Always-on rules

## Personal Rules with AGENTS.local.md

[Personal Rules with AGENTS.local.md](https://docs.devin.ai/cli/extensibility/rules#personal-rules-with-agents-local-md):

1. Personal instructions that shouldn't be shared with collaborators (such as preferred working style, testing habits, or review preferences).
1. Create an `AGENTS.local.md` file next to your `AGENTS.md`.
1. Loaded alongside `AGENTS.md` with the same always-on behavior.
1. ⚠️ Add it to your `.gitignore` so it stays local.

## Rules Loading Procedure

[Rules/agents files can exist at multiple levels in your project (not just the root)](https://docs.devin.ai/cli/extensibility/rules#supported-file-names):
* Files at the workspace root are loaded at session start.
* Files in subdirectories are discovered lazily when the agent accesses files in that directory, keeping the context focused on the relevant part of the codebase.
