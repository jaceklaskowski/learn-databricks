
## Runnable

📖 [View in docs](https://reference.langchain.com/python/langchain-core/runnables/base/Runnable)

A unit of work that can be invoked, batched, streamed, transformed and composed.

Programs created using LCEL and LangChain `Runnable` objects inherently support synchronous, asynchronous, batch, and streaming operations.

* **Async** for scalability and higher concurrent loads.
* **Batch** for parallel input processing.
* **Streaming** (of intermediate outputs, as they're being generated) for more responsive UX.

### Key Methods

- `invoke`/`ainvoke`: Transforms a single input into an output.
- `batch`/`abatch`: Efficiently transforms multiple inputs into outputs.
- `stream`/`astream`: Streams output from a single input as it's produced.
- `astream_log`: Streams output and selected intermediate results from an
    input.

Built-in optimizations:

- **Batch**: By default, batch runs invoke() in parallel using a thread pool
    executor. Override to optimize batching.

- **Async**: Methods with `'a'` prefix are asynchronous. By default, they execute
    the sync counterpart using asyncio's thread pool.
    Override for native async.

All methods accept an optional config argument, which can be used to configure
execution, add tags and metadata for tracing and debugging etc.

Runnables expose schematic information about their input, output and config via
the `input_schema` property, the `output_schema` property and `config_schema`
method.


### Composition

`Runnable` objects can be composed together to create chains in a declarative way.

Any chain constructed this way will automatically have sync, async, batch, and
streaming support.

The main composition primitives are `RunnableSequence` and `RunnableParallel`.

- **`RunnableParallel`** invokes runnables concurrently, providing the same input
to each. Construct it using a dict literal within a sequence or by passing a
dict to `RunnableParallel`.

- **`RunnableSequence`** invokes a series of runnables sequentially, with
one Runnable's output serving as the next's input. Construct using
the `|` operator or by passing a list of runnables to `RunnableSequence`.

### RunnableLambda

````python
from langchain_core.runnables import RunnableLambda

# A RunnableSequence constructed using the `|` operator
sequence = RunnableLambda(lambda x: x + 1) | RunnableLambda(lambda x: x * 2)

print(f"sequence.invoke(1): {sequence.invoke(1)}")
print(f"sequence.batch([1, 2, 3]): {sequence.batch([1, 2, 3])}")
````

````python
from langchain_core.runnables import RunnableLambda

# A sequence that contains a RunnableParallel constructed using a dict literal
sequence = RunnableLambda(lambda x: x + 1) | {
    "mul_2": RunnableLambda(lambda x: x * 2),
    "mul_5": RunnableLambda(lambda x: x * 5),
}
print(f"sequence.invoke(1): {sequence.invoke(1)}")
````

### RunnableBranch

**RunnableBranch** that selects which branch to run based on a condition.
Initialized with a list of `(condition, Runnable)` pairs and the default branch.

````python
from langchain_core.runnables import RunnableBranch

branch = RunnableBranch(
    (lambda x: isinstance(x, str), lambda x: x.upper()),
    (lambda x: isinstance(x, int), lambda x: x + 1),
    (lambda x: isinstance(x, float), lambda x: x * 2),
    lambda x: "goodbye",
)

assert branch.invoke("hello") == "HELLO"
assert branch.invoke(None) == "goodbye"
````

### RouterRunnable

**RouterRunnable** that routes to a set of `Runnable` based on `Input['key']`.

````python
from langchain_core.runnables.router import RouterRunnable
from langchain_core.runnables import RunnableLambda

add = RunnableLambda(func=lambda x: x + 1)
square = RunnableLambda(func=lambda x: x**2)

router = RouterRunnable(runnables={"add": add, "square": square})

assert router.invoke({"key": "square", "input": 3}) == 9
````

### RunnableAssign

Assigns key-value pairs to `dict[str, Any]` inputs.

````python
# This is a RunnableAssign
from langchain_core.runnables.passthrough import (
    RunnableAssign,
    RunnableParallel,
)
from langchain_core.runnables.base import RunnableLambda


def add_ten(x: dict[str, int]) -> dict[str, int]:
    return {"added": x["input"] + 10}


mapper = RunnableParallel(
    {
        "add_step": RunnableLambda(add_ten),
    }
)

runnable_assign = RunnableAssign(mapper)

# Synchronous example
assert runnable_assign.invoke({"input": 5}) == {'input': 5, 'add_step': {'added': 15}}

# Asynchronous example
actual = await runnable_assign.ainvoke({"input": 5})
expected = {'input': 5, 'add_step': {'added': 15}}

assert actual == expected
````

## @chain

`langchain_core.runnables.base.chain` decorator creates a `RunnableLambda` over the wrapped `func`.

````python
from langchain_core.runnables import chain
from langchain_core.prompts import PromptTemplate
from langchain_openai import OpenAI


@chain
def my_func(fields):
    prompt = PromptTemplate("Hello, {name}!")
    model = OpenAI()
    formatted = prompt.invoke(**fields)

    for chunk in model.stream(formatted):
        yield chunk
````

# Why LangChain on Databricks? Why not Databricks alone?

I asked Copilot Chat the following questions and received this response:


A good way to think about it is:

**Databricks and LangChain solve different problems.**

- **Databricks** = the platform (compute, data, vector search, model serving, governance, monitoring, agents).
- **LangChain** = the application orchestration framework (chains, agents, tools, memory, prompt workflows, multi-step reasoning).

So the question is rarely **"Databricks or LangChain?"**. More often it's **"Databricks with or without LangChain?"**

## What Databricks already provides

Databricks can build many GenAI applications without LangChain:

- Data storage and processing (Delta Lake)
- Vector Search
- Model Serving
- MLflow model lifecycle
- Mosaic AI Agent Framework
- Evaluation and monitoring
- Unity Catalog governance
- RAG pipelines

For a straightforward RAG solution, Databricks alone is often enough:

```text
User Query
    ↓
Databricks Vector Search
    ↓
Retrieved Documents
    ↓
LLM Endpoint
    ↓
Answer
```

This architecture is simpler, easier to operate, and has fewer dependencies.

---

## What LangChain adds

LangChain becomes useful when orchestration gets complex.

### 1. Multi-step workflows

Instead of:

```text
Question → Search → Answer
```

you may need:

```text
Question
   ↓
Classify intent
   ↓
Choose tool
   ↓
Query SQL
   ↓
Query Vector DB
   ↓
Call API
   ↓
Combine results
   ↓
Answer
```

LangChain provides abstractions for these flows.

---

### 2. Tool calling

Suppose an assistant can:

- Query Databricks SQL
- Search documents
- Check Jira
- Call ServiceNow
- Query REST APIs

LangChain makes tool registration and routing easier.

Example:

```python
agent = create_agent(
    model=model,
    tools=[
        sql_tool,
        vector_search_tool,
        jira_tool
    ]
)
```

---

### 3. Vendor portability

Without LangChain:

```text
Application
    ↓
Databricks endpoint
```

With LangChain:

```text
Application
    ↓
LangChain abstraction
    ↓
OpenAI / Azure OpenAI / Databricks / Anthropic
```

If your organization frequently changes models, LangChain reduces application rewrites.

---

### 4. Agent frameworks

LangChain was designed around:

- Agents
- Tool use
- Planning
- Multi-agent systems
- Graph workflows (via LangGraph)

Databricks provides agent capabilities too, but LangGraph is often considered more flexible for highly customized agent architectures.

---

## Why some teams avoid LangChain

Many production teams intentionally avoid it because:

### Added complexity

You introduce another layer:

```text
Application
    ↓
LangChain
    ↓
Databricks
    ↓
Model
```

More libraries, upgrades, debugging, and compatibility concerns.

### Frequent API changes

LangChain historically evolved rapidly, requiring code updates across versions.

### Latency overhead

Very complex chains can increase response time versus a direct Databricks implementation.

### Databricks already covers many needs

With Mosaic AI and Agent Framework, some LangChain functionality is now natively available in Databricks.

---

## Typical recommendation

### Use Databricks alone when:

- Simple RAG
- Enterprise search
- Chat over documents
- SQL + LLM workflows
- Need strong governance and simplicity

### Add LangChain when:

- Complex agent workflows
- Many external tools
- Multi-step reasoning pipelines
- Need portability across LLM vendors
- Already have a large LangChain ecosystem

---

## As a Databricks Engineer perspective

For most enterprise projects today, I would start with **Databricks-native capabilities first**. If the architecture grows into sophisticated agent orchestration, then introduce **LangChain/LangGraph selectively**.

A common modern stack is:

```text
Databricks
├── Delta Lake
├── Unity Catalog
├── Vector Search
├── Model Serving
├── MLflow
└── Agent Evaluation

LangGraph/LangChain
└── Agent orchestration layer

Application
└── UI/API
```

So the real answer is:

> **Databricks provides the AI platform; LangChain provides orchestration.**
> For simple RAG, Databricks alone is often sufficient. For complex agentic workflows, LangChain/LangGraph can add significant value on top of Databricks.
