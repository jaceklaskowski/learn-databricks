## PromptTemplate

[PromptTemplate](https://reference.langchain.com/python/langchain-core/prompts/prompt/PromptTemplate)

* A prompt template for a language model.
* A prompt template consists of a string template.
* A prompt template accepts a set of parameters from the user that can be used to generate a prompt for a language model.
* The template can be formatted using the following syntax:
    * f-strings (default)
    * jinja2
    * mustache

```python
from langchain_core.prompts import PromptTemplate
help(PromptTemplate.from_template)
```

```python
from langchain_core.prompts import PromptTemplate

# Instantiation using from_template (recommended)
prompt = PromptTemplate.from_template("Say {foo}")
prompt.format(foo="bar")

print(f"prompt: {prompt}")

# Instantiation using initializer
prompt = PromptTemplate(template="Say {foo}")
print(f"prompt: {prompt}")
```

```python
from langchain_core.prompts import PromptTemplate
help(PromptTemplate.from_file)
```

```python
from langchain_core.prompts import PromptTemplate

# Create a prompt template text file for demo completeness
with open("weather_prompt_template.txt", "w") as f:
    f.write("What's the weather like in {city} now?")

# Instantiation using from_template (recommended)
prompt = PromptTemplate.from_file(template_file="weather_prompt_template.txt")
print(f"prompt: {prompt}")
```

## ChatDatabricks

```python
from databricks_langchain import ChatDatabricks
help(ChatDatabricks)
```

```python
from databricks_langchain import ChatDatabricks

# Replace with your Databricks model endpoint name
LLM_ENDPOINT_NAME = "databricks-meta-llama-3-1-70b-instruct"

llm = ChatDatabricks(endpoint=LLM_ENDPOINT_NAME, temperature=0.1)

response = llm.invoke(prompt.format(city="Warsaw, Poland"))
print(response)
```

```python
from langchain_core.output_parsers import StrOutputParser

output_parser = StrOutputParser()
```

```python
llm_chain = prompt | llm | output_parser
```

```python
response = llm_chain.invoke(input={"city": "Warsaw, Poland"})
print(response)
```
