
# Demo: MLflow with PostgreSQL Backend Store

## Start PostgreSQL

<br>

```shell
/opt/homebrew/opt/postgresql@18/bin/postgres -D /opt/homebrew/var/postgresql@18
```


## Set Up Database

<br>

```shell
psql -U postgres
```

<br>

```shell
createuser user
createdb mlflow
psql -d mlflow -c "GRANT CREATE ON SCHEMA public TO \"user\";"
psql -U user --dbname=mlflow
```

<br>

```shell
## List the names of all the databases in PostgreSQL
mlflow=> \l
```

## Start MLflow Server


[\[Solved\] Port 5000 Used by Control Center in macOS Monterey](https://nono.ma/port-5000-used-by-control-center-in-macos-controlce):

> In macOS Tahoe, you can find AirPlay Receiver toggle in System Settings > AirDrop & Handoff.


```shell
MLFLOW_DISABLE_TELEMETRY=1 \
uvx --from mlflow --with psycopg2-binary mlflow server \
  --port 5555 \
  --backend-store-uri postgresql://user@localhost/mlflow \
  --default-artifact-root s3://mlflow-artifacts \
  --enable-workspaces
```

## MLflow UI

http://127.0.0.1:5555


```shell
MLFLOW_TRACKING_URI=http://127.0.0.1:5555/ \
uvx --from "mlflow-skinny" mlflow doctor
```

## MLflow CLI

### Create Experiment


```shell
MLFLOW_TRACKING_URI=http://127.0.0.1:5555/ \
uvx --from "mlflow-skinny" \
mlflow experiments create --experiment-name "Jacek's Great Experiment"
```

```shell
MLFLOW_TRACKING_URI=http://127.0.0.1:5555/ \
uvx --from "mlflow-skinny" \
mlflow experiments get --experiment-id 1
```
