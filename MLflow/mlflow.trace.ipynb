
# @mlflow.trace


A [Python decorator](https://realpython.com/primer-on-python-decorators/) that creates a new span for the decorated function.

<br>

```py
@mlflow.trace
def my_fun(x: int, y: str) -> str:
    return y * x
```

`@mlflow.trace` creates a span for the scope of the decorated function.

## Demo

````python
import mlflow
from mlflow.entities import SpanType


@mlflow.trace
def repeat_text(text: str, how_many: int) -> str:
    return text * how_many

@mlflow.trace
def uppercase_text(text: str) -> str:
    return text.upper()

@mlflow.trace(
    name="main agent",
    span_type=SpanType.TOOL,
)
def repeat_uppercase_text(text: str, how_many: int = 3) -> str:
    return uppercase_text(repeat_text(text, how_many))
````

````python
repeat_uppercase_text("MLflow rocks!")
````


## Arguments


### span_type

The type of the span (a string or one of the values of `mlflow.entities.SpanType`).

Spans have got their own icons in the UI


### name

The name of the span.

Unless provided, the name of the function will be used.
