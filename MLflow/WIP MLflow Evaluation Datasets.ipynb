An **evaluation dataset** in MLflow is a selected set of example inputs — either **labeled** (with known expected outputs) or **unlabeled** (without ground-truth answers).

You use evaluation datasets to systematically test and improve a GenAI application.

MLflow evaluation datasets are stored in Unity Catalog (built-in versioning, lineage, sharing, and governance).

An evaluation dataset is attached to an MLflow experiment.

## Requirements

1. `CREATE TABLE` permission on a Unity Catalog schema.
1. [MLflow experiment](https://docs.databricks.com/aws/en/mlflow3/genai/getting-started/connect-environment#create-expt) (the eval dataset will be attached to).

## Data Sources

* Existing traces
* An existing dataset, or directly entered examples.
* Synthetic data (Databricks can automatically generate a representative evaluation set from your documents)

Datasets can be passed as pandas DataFrames, Spark DataFrames, or Lists of Dictionaries.

See [MLflow evaluation examples for GenAI](https://docs.databricks.com/aws/en/mlflow3/genai/eval-monitor/eval-examples)

## Format

An **Evaluation Dataset** is a tabular dataset with two columns:

1. `Inputs`
1. `Expectations`


### Inputs

`{"messages":[{"role":"user","content":"Hello"}]}`

### Expectations

`{"guidelines":["The response must be professional"]}`
