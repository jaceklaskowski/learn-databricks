## Exam Coverage

Candidates are tested on the following topics:

1. Developing Code for Data Processing using Python and SQL (batch and streaming workloads)
1. Data Ingestion & Acquisition
1. Data Transformation, Cleansing, and Quality
1. Data Sharing and Federation
1. Monitoring and Alerting
1. Cost & Performance Optimisation
1. Ensuring Data Security and Compliance
1. Data Governance
1. Debugging and Deploying
1. Data Modelling
1. Orchestrating workflows
1. Leveraging DevOps and CI/CD
1. Deploying with tools like the Databricks CLI, REST API, and Asset Bundles.

## Core Platform Features

Successful candidates demonstrate expertise across core platform features:

* Python and SQL
* Delta Lake
* Unity Catalog
* Auto Loader
* Lakeflow Spark Declarative Pipelines
* Databricks Compute (including serverless)
* Lakeflow Jobs
* Medallion Architecture
