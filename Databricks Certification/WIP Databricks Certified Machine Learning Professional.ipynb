# MLOps

**Machine Learning Operations (MLOps)** = from a business problem to deployment and monitoring of machine learning models.

* Data Collection
* Data Preprocessing / Feature Engineering
* Model Training
* Model Evaluation
* Model Deployment
* Model Monitoring
