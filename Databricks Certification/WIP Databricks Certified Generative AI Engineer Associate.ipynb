# Exam Assessesment

Assesses ability to design and implement LLM-enabled solutions using Databricks:

* [Databricks Vector Search](https://docs.databricks.com/aws/en/vector-search/vector-search) for semantic similarity searches
    * Create and query a Vector Search index
    * Resources needed to serve features
* Model Serving for deploying models and solutions
    * how to serve an LLM application that leverages Foundation Model APIs
    * Batch inference workloads with `ai_query()`
* MLflow for managing solution lifecycle
    * Register models to Unity Catalog using MLflow
* Unity Catalog for data governance
