# Fine-Grained Access Control

[Row filters and column masks](https://docs.databricks.com/aws/en/data-governance/unity-catalog/filters-and-masks)

## Row-Level Security

**Row Filtering** is used to apply a filter (`CREATE FUNCTION`) to a table using `ALTER TABLE`.

<br>

```sql
ALTER TABLE ...
SET ROW FILTER ... ON ...
```

## Column-Level Masking

**Column Masking** is used to apply a masking function to a table column.

<br>

```sql
ALTER TABLE ...
ALTER COLUMN ...
SET MASK ...
```
